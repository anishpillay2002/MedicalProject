```python
# Decision Tree Regression

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import qgrid
```

```python
df = pd.read_excel('ABPI RESEARCH FOR CALCULATION-2.xlsx', sheet_name='SHEET 1')
pd.set_option('display.max_columns', 999)
print("Column headings:")
df
```

```python
def CleaningUp(df):

    #Deleting First row as it contains NaN
    df = df.drop([0])
    # Replacing PostMenoPausal to NotApp for Male patients and No for female if its set as '-'
    for index, row in df.iterrows():
        if df.loc[index, "SEX"] == "M":
            df.loc[index, "POSTMENOPAUSAL"] = "NotApp"
        elif df.loc[index, "SEX"] == "F" and df.loc[index, "POSTMENOPAUSAL"] == "-":
            df.loc[index, "POSTMENOPAUSAL"] = "NO"
    
    
    ### Binning Tobacco Consumption Data
    plt.figure();
    plt.title('Histogram Tobacco Consumption(RawData)')
    df['TOBACCO CONSUMPTION (YRS)'].plot(kind='hist')
    df['TOBACCO CONSUMPTION (YRS)'].describe()
    
    # Binning Tobacco consumption data and storing it in datafrome
    cut_labels = ['0', '1', '2']
    df['Binned_TOBACCO CONSUMPTION (YRS)'] = pd.cut(df['TOBACCO CONSUMPTION (YRS)'], bins=3,labels = cut_labels)
    
    
    # Checking size of each bin
    print('\n\nBin sizes of Tobacco Consumption')
    print(df['Binned_TOBACCO CONSUMPTION (YRS)'].value_counts())
    
    ### Binning Duratino of Diabetes Data
    plt.figure();
    plt.title('Histogram DURATION OF DIABETES(RawData)')
    df['DURATION OF DIABETES (YRS)'].plot(kind='hist')
    df['DURATION OF DIABETES (YRS)'].describe()
    
    # Binning DURATION OF DIABETES (YRS) and storing in dataframe
    cut_labels = ['0', '1', '2']
    df['Binned_DURATION OF DIABETES (YRS)'] = pd.qcut(df['DURATION OF DIABETES (YRS)'], q=3, labels = cut_labels)
    
    print('\n\nBin sizes of DURATION OF DIABETES')
    print(df['Binned_DURATION OF DIABETES (YRS)'].value_counts())
    
    ### Binning Age
    plt.figure();
    plt.title('Histogram Age(RawData)')
    df['AGE'].plot(kind='hist')
    df['AGE'].describe()
    
    # Binning AGE and storing in dataframe
    cut_labels = ['0', '1', '2','3','4']
    df['Binned_AGE'] = pd.cut(df['AGE'], bins=5,labels = cut_labels)
    
    print('\n\nBin sizes of Age')
    print(df['Binned_AGE'].value_counts())
    
    ### Deleting empty columns
    df = df.drop(columns = ' ')
    return df
```

```python
df = CleaningUp(df)
```

```python
# Left Leg
X_left = df.iloc[:, [2,3,5,6,7,8,9,10,11,12,13,22,25,26,27]].values
y_left = df.iloc[:, 24].values

# Right Leg
X_right = df.iloc[:, [2,3,5,6,7,8,9,10,11,12,13,21,25,26,27]].values
y_right = df.iloc[:, 23].values
```

```python
# Label Encoder
from sklearn.preprocessing import LabelEncoder 
from sklearn.preprocessing import OneHotEncoder 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

def ModifyingVariables(X,y):
    le = LabelEncoder() 
  
    # Encoding all the string independent variables(X)
    X[:, 0] = le.fit_transform(X[:, 0])
    X[:, 1] = le.fit_transform(X[:, 1])
    X[:, -1] = le.fit_transform(X[:, -1])
    X[:, -2] = le.fit_transform(X[:, -2])
    X[:, -3] = le.fit_transform(X[:, -3])
    
    # Replacing values> 0 to 1 in dependent variable(y)
    y = np.where(y>0, 1, y) 
    
    # One hot encoder

    # creating one hot encoder object by default 
    # entire data passed is one hot encoded 
    onehotencoder = OneHotEncoder() 
    X = onehotencoder.fit_transform(X).toarray() 
    
    return X,y

def SplittingDataset(X,y):
    # Splitting the dataset into the Training set and Test set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
    X_train, X_val, y_train, y_val  = train_test_split(X_train, y_train, test_size=0.2, random_state=1)
    
    # Feature Scaling

    sc_X = StandardScaler()
    X_train = sc_X.fit_transform(X_train)
    X_val = sc_X.transform(X_val)
    X_test = sc_X.transform(X_test)
    
    return X_train,X_val,X_test,y_train,y_val,y_test
    
    
def FittingModel(X_train,y_train):
    # Fitting K-Neighbors Classifier to training set
    from sklearn.neighbors import KNeighborsClassifier
    classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p=2);
    classifier.fit(X_train,y_train)
    return classifier

def ApplyPCA(X_train,X_test):
    # Applying PCA
    pca = PCA(n_components = 4)
    X_train = pca.fit_transform(X_train)
    X_test = pca.transform(X_test)
    explained_variance = pca.explained_variance_ratio_
    return explained_variance
```

```python
# Applying PCA to check
tmp_X_train, tmp_X_test, tmp_y_train, tmp_y_test = train_test_split(X_left, y_left, test_size=0.2, random_state=1)
explained_variance = ApplyPCA(tmp_X_train,tmp_X_test)
print(explained_variance)
```

# Left Leg

```python
## For Left Leg
X,y = ModifyingVariables(X_left,y_left);
X_train,X_val,X_test,y_train,y_val,y_test = SplittingDataset(X,y);
classifier = FittingModel(X_train,y_train);
```

```python
### For Left Leg
## Cross Validation Data
# Predict 
y_val_pred = classifier.predict(X_val);

# Confusion Matrix
print('\n\nCross Validation - Confusion Matrix')
pd.crosstab(y_val, y_val_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
### For Left Leg
## Cross Validation Data
# Predict 
y_test_pred = classifier.predict(X_test);
print(y_test_pred)

# Confusion Matrix
print('\n\nTest Dataset - Confusion Matrix- Left Leg')
pd.crosstab(y_test, y_test_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

# Right Leg

```python
## For Right Leg
X,y = ModifyingVariables(X_right,y_right);
X_train,X_val,X_test,y_train,y_val,y_test = SplittingDataset(X,y);
classifier = FittingModel(X_train,y_train);
```

```python
### For Right Leg
## Cross Validation Data
# Predict 
y_val_pred = classifier.predict(X_val);

# Confusion Matrix
print('\n\nCross Validation - Confusion Matrix- Right Leg')
pd.crosstab(y_val, y_val_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
### For Right Leg
## Cross Validation Data
# Predict 
y_test_pred = classifier.predict(X_test);
print(y_test_pred)

# Confusion Matrix
print('\n\nTest Dataset - Confusion Matrix- Right Leg')
pd.crosstab(y_test, y_test_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```
